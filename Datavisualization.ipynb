# univariant continous data a contnious - histogram in used to see analysis of univariant cotinous data

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
""" mtcars data columns
mpg =miles per galon
cyl = cyinder
disp= total volume of cyinder in liters or cubic
hp = horsepower [ more horsepower = high power of engine]
drat = rear axle ratio [ affect car acceleratin and fuel efficeincy]
wt = wieght of a car
qsec = perfomance matric of car's acceleration
vs = confifuration of engine is v = 0 v engine or vs=1 straight engine
am = transmission = 1 manual , 0 =automatic
gear = nuer of forward gear
carb = number of carbeurators which mix fuel with air before adding it to engine"""
```

```python
mtcars_data=pd.read_csv('mtcars.csv')
mtcars_data

```

```python
mtcars_data.head()
```

```python
mtcars_data.shape
```

```python
plt.figure(figsize=(5,5))
plt.hist(x='mpg',data=mtcars_data)   # willshow arays
plt.xlabel('mpg')
plt.ylabel('frequency')
plt.title('mpg distance')
plt.show()
```

## univariant - a discrete - bar chart is used to see anlaysis of univariant discrete value

```python
plt.figure(figure=(10,5))
sns.countplot(x='gear',data=mtcars_data)
plt.show()
```

## 1 bivariatn -1 continous 1 disrete - bar chart is used to see analysis of bivraiant data

```python
mtcars_data['cyl'].value_counts()
```

```python
# x is discrete , yis continous

sns.barplot(x='gear',y='mpg',data=mtcars_data , hue='gear') # hue - colour on the basis of discrete
plt.title('MPG vs gears')
plt.show()
```


## multivaraint - boxplot,scatterplot is used to see anaylsis of mutivariannt


```python
sns.boxplot(x='cyl',y='mpg',data=mtcars_data )
plt.show()
```

```python
# Scatter chart -3  continous
sns.scatterplot(x='hp' , y='mpg' , data=mtcars_data, hue='hp')
```

```python
sns.scatterplot(x='disp' , y='hp',data=mtcars_data ,hue='hp')
```
